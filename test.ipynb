```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x, y):
    z = np.column_stack([x.ravel(), y.ravel()])
    print(z)
    return (0.1*st.multivariate_normal([0,0], 1*np.eye(2)).pdf(z) +
            0.4*st.multivariate_normal([3,3], 2*np.eye(2)).pdf(z) +
            0.5*st.multivariate_normal([0,5], 3*np.eye(2)).pdf(z))
```

```python
z = f(np.arange(3), np.arange(3))
print(z)
```

```python
x = np.linspace(-5, 10, 20)
print(x)
z = np.column_stack([x.ravel(), x.ravel()])
d = (0.1*st.multivariate_normal([0,0], 1*np.eye(2)).pdf(z) +
            0.4*st.multivariate_normal([3,3], 2*np.eye(2)).pdf(z) +
            0.5*st.multivariate_normal([0,5], 3*np.eye(2)).pdf(z))
print(d.shape)
```

```python
s = 200
x = np.linspace(-5, 10, s)
y = np.linspace(-5, 10, s)
X, Y = np.meshgrid(x, y)
print(X.shape)
print(Y.shape)
Z = np.reshape(f(X, Y), (s, s))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
plt.title('Gaussian Mixture Model')
```
